# Riemann-$\zeta$ function and its zeros

The function $$ \zeta(s) = \sum_{n=1}^\infty \frac{1}{n^s}  = \prod_{p \text{ prime}} \left(1 + \frac{1}{p^s} + \frac{1}{p^{2s}} + \frac{1}{p^{3s}} + \cdots\right) = \prod_{p \text{ prime}} \left(1 - \frac{1}{p^s}\right)^{-1}$$ is defined for $\operatorname{Re}(s)>1$. 

On top of each factor <i>at primes</i> $p$, adding also the factor "<i>at infinity</i>" we arrive at the completed zeta function:
$$
    \zeta^*(s) = \zeta(s) \Gamma\left(\frac{s}{2}\right) \pi^{-\frac{s}{2}} 
$$

The completed zeta function has a meromorphic continuation to all $s \in \mathbb{C}$ with a simple pole at $s = 1$.

Here $\Gamma(s)$ is the well understood complex function that extends the factorial function. It is defined by
$$
    \Gamma(s) = \int_0^\infty e^{-t} t^s \tfrac{\mathrm{d} t}{t} \qquad \qquad \text{ for } \operatorname{Re}(s)>0,
$$
and satisfies $\Gamma(s) = s\Gamma(s-1)$ with $\Gamma(1) = 1$. Therefore $\Gamma(n) = (n-1)!$.

## Theta and Riemann zeta.

Given the theta function defined on $\operatorname{Im}(z)>0$, 
$$
    \theta(z) = \sum_{n \in \mathbb{Z}} e^{\pi i n^2 z} 
$$
We have that
$$
    \frac{1}{2}\left(\theta(iy) - 1\right) = \sum_{n=1}^\infty e^{-\pi n^2 y}.
$$
Thus we can calculate as follows,
$$
\begin{align}
    \int_0^\infty \frac{1}{2}\left(\theta(iy) - 1\right) y^{\frac{s}{2}} \frac{\mathrm{d}y}{y} &= \int_0^\infty \sum_{n=1}^\infty e^{-\pi n^2 y} y^{\frac{s}{2}} \tfrac{\mathrm{d} y}{y} \\
    &= \sum_{n=1}^\infty \int_0^\infty  e^{-\pi n^2 y} y^{\frac{s}{2}} \tfrac{\mathrm{d} y}{y}  \qquad  \left(\text{ change } y \mapsto \tfrac{y}{\pi n^2}\right)\\
    &= \sum_{n=1}^\infty \frac{1}{n^s} \pi^{-\frac{s}{2}} \int_{0}^\infty e^{-y} y^{\frac{s}{2}} \tfrac{\mathrm{d}y}{y} \\
    &= \zeta(s) \pi^{-\frac{s}{2}} \Gamma\left(\frac{s}{2}\right)
\end{align}
$$

## The functional equation and analytic continuation

The completed zeta function has a meromorphic continuation to all $s \in \mathbb{C}$ and satisfies the functional equation
$$ 
    \zeta^*(s) = \zeta^*(1-s).
$$
The only pole of $\zeta(s)$ is at $s = 1$ with residue $1$ (the completed zeta function has two poles, one more at $s = 0$ by symmetry coming from $\Gamma(s/2)$. 


The theta function satisfies functional equations of the form $\theta(z + 2) = \theta(z)$ and $$\theta\left(\frac{-1}{z}\right) = \sqrt{\frac{z}{i}}\theta(z).$$

For $z = iy$ this follows from the Poisson summation formula which states that for a function $f:\mathbb{R} \longrightarrow \mathbb{C}$
$$
\sum_{n \in \mathbb{Z}} f(n) = \sum_{m \in \mathbb{Z}} \hat{f}(m),
$$
as long as both sides makes sense and converge.

For $f(x) = e^{-\pi x^2 y}$ we have that its Fourier transform is $\hat{f}(\xi) = \frac{1}{\sqrt{y}}e^{-\pi x^2 /y}$, and therefore 
$$
   \theta(iy) =  \sum_{n\in \mathbb{Z}} e^{-\pi n^2 y} = \frac{1}{\sqrt{y}} \sum_{m \in \mathbb{Z}} e^{-\pi m^2/y} = \frac{1}{\sqrt{y}}\theta\left(-\frac{1}{iy}\right)
$$
which means
$$
    \tfrac{1}{2}\left(\theta\left(-\tfrac{1}{iy}\right) - 1\right) = \tfrac{1}{2}\left(\sqrt{y}\theta(iy) -1\right) = \tfrac{\sqrt{y}}{2}\left(\theta(iy)-1\right) + \frac{\sqrt{y}}{2} - \frac{1}{2}.
$$

Putting this in the above integral we have that
$$
\begin{align}
    \zeta(s) \Gamma\left(\tfrac{s}{2}\right) \pi^{-\frac{s}{2}} &=  \int_{0}^\infty \tfrac{1}{2}(\theta(iy)-1) y^{\frac s2} \tfrac{\mathrm{d}y}{y} \\
    &=\left(\int_0^1 +  \int_{1}^\infty\right) \tfrac{1}{2}(\theta(iy)-1) y^{\frac s2} \tfrac{\mathrm{d}y}{y} \\
    &= \int_{1}^\infty \tfrac{1}{2}\left(\theta\left(\tfrac{1}{iy}\right)-1\right) y^{-\frac s2} \tfrac{\mathrm{d}y}{y} + \int_{1}^\infty \tfrac{1}{2}(\theta(iy)-1) y^{\frac s2} \tfrac{\mathrm{d}y}{y} \quad \text{ sending } y \mapsto \tfrac 1y\\
    &= \int_{1}^\infty \tfrac{1}{2}\left(\theta(iy) - 1\right)) \left(y^{\frac{1-s}{2}} + y^{\frac{s}{2}} \right) \tfrac{\mathrm{d} y}{y} + \frac{1}{2} \int_{1}^\infty \left(y^{\tfrac{-s-1}{2}} - y^{-\frac{s}{2}}\right)\tfrac{\mathrm{d}y}{y}\\
    &= \int_{1}^\infty \tfrac{1}{2}\left(\theta(iy) - 1\right)) \left(y^{\frac{1-s}{2}}+ y^{\frac{s}{2}} \right) \tfrac{\mathrm{d} y}{y}  - \frac{1}{1-s} - \frac{1}{s}
\end{align}
$$

This is a symmetric formula under $s \leftrightarrow 1-s$ and converges for all $s \in \mathbb{C}$.

We conclude that the completed zeta function
$$\zeta^*(s) = \zeta(s)\Gamma\left(\tfrac{s}{2}\right) \pi^{-\frac{s}{2}}$$
has simple poles only at $s = 0$ and $s = 1 $ with residues equal to $-1$ and $1$ respectively. It also satisfies the functional equation
$$
\zeta^*(s) = \zeta^*(1-s).
$$

## Calculating the zeta function from the above formula


Note that we have a quickly convergent integral, in fact we can express this last line as
$$
\begin{align*}
    \zeta^*(s)  +\frac{1}{s} - \frac{1}{s-1} &= \int_1^\infty \frac{1}{2}\left(\theta(iy) - 1\right) y^{\frac{s}{2}} \tfrac{\mathrm{d}y}{y}\\
    &= \int_1^\infty \sum_{n=1}^\infty e^{-\pi n^2 y} \left(y^{\frac{s}{2}} + y^\frac{1-s}{2}\right) \tfrac{\mathrm{d}y}{y}\\
    &= \pi^{-\frac{s}{2}} \sum_{n = 1}^\infty \frac{1}{n^s} \int_{\pi n^2}^\infty e^{-y} y^{\frac{s}{2}} \frac{\mathrm{d} y}{y}  + \pi^{-\frac{1-s}{2}} \sum_{n=1}^\infty \frac{1}{n^{1-s} }  \int_{\pi n^2}^\infty e^{-y} y^{\frac{1-s}{2}} \frac{\mathrm{d} y}{y} \\
    &= \pi^{-\frac{s}{2}} \sum_{n = 1}^\infty \frac{1}{n^s} \Gamma\left(\frac s 2, \pi n^2\right) + \pi^{-\frac{1-s}{2}} \sum_{n=1}^\infty \frac{1}{n^{1-s} }  \Gamma\left(\frac{1-s}{2}, \pi n^2\right) 
\end{align*}
$$
 
The incomplete gamma functions decay very quickly with $n$. So the right hand side is a quickly convergent $L$-function. That is good. Thus the $\zeta^*$ function can be calculated very efficiently. 

The issue is that for high values of $t$ in $s = \sigma + it$ the functions $\Gamma(\tfrac{\sigma + it}{2})$ and $\Gamma(\tfrac{\sigma + it}{2})$ both decay very quickly, as fast as $O(e^{-\pi t/2})$. Computer arithmetic is not very good when dividing two very small numbers. Thus we need to compute original numbers to even higher precision.

```python
CC= ComplexField(prec=100)
RR = RealField(prec = 100)
RR
```

```python
from scipy.integrate import quad

def zetastar_from_theta(s, terms):
    total1 = 0
    total2 = 0
    sigma = RR(s.real_part())
    t = RR(s.imag_part())
    for n in range(1,terms):
        nn = CC(n)*CC(n)
        cs = cos(t*log(RR(n)))
        sn = sin(t*log(RR(n)))
        total1 += gamma_inc(CC(s)/2, CC(pi)*nn)*(RR(n)**(-sigma))*(cs - CC(i)*sn)
        total2 += gamma_inc(CC((1-s)/2), CC(pi)*nn)*(RR(n)**(sigma-1))*(cs + CC(i)*sn)
    total1 *= CC(pi)**CC(-s/2)
    total2 *= CC(pi)**CC((s-1)/2)
    return total1 + total2 - 1/(CC(s)) - 1/CC(1-s)


def zeta_from_theta(s,terms):
    extra = gamma(CC(s)/2)*(CC(pi)**(-CC(s/2)))
    return zetastar_from_theta(s, terms)/extra

print('now: \t\t',zetastar_from_theta(0.5+ 12*i, 10),'\npreviously:\t',zetastar(1/2+12*i))
```

## Newton Method

One way to find zeros of a complex (or real) function $f$ is to apply the newton iteration to a point. When it converges, it has quadratic convergence. As with real valued functions the iterations are simply defined as     
$$
    z_{n+1} = z_n - \frac{f(z_n)}{f'(z_n)}
$$
where $z_0$ is a point close enough to a root of $f$.

<img src="images/newtonmethod.gif" width=500px/> 
<center> <font size =1> Newton Method. Via wikimedia commons </font> </center>

```python
CC = ComplexField()

def newton_method(init, f, df = 'numerical', tol = 1e-8):
    z = CC(init)
    if df =='numerical':
        df = numerical_derivative(f, 1e-6)
    while abs(f(CC(z))) > tol:
        z = z - CC(f(z)/df(z))
    return CC(z)
        
def numerical_derivative(f, h):
    df = lambda p: (f(CC(p + h)) - f(CC(p-h)))/(CC(2*h))
    return df
```

## Gradient Descent

One can also reach the minimum of the function $|\zeta(s)|$ by gradient descent algorithm.

The minima are the zeros of $\zeta(s)$. 

<img src="./images/Gradient_descent.gif" width = 500px/>
<center> <font size = 1> Image from wikimedia commons</font></center>



```python
RR = RealField(prec = 150)
CC = ComplexField(prec = 150)

def gradient_descent(f, init, atol = 1e-8):
    """
    f complex function finds minimum of |f| starting with z = init. 
    """
    z = CC(init)
    eta = RR(0.5)
    while abs(f(z)) > atol:
        df = (f(z + CC(1e-6)) - f(z - CC(1e-6)))*CC(1e6)/2
        w = z - eta*f(z)*conjugate(df)/abs(f(z))
        if abs(f(w)) > abs(f(z)):
            eta = eta*RR(0.5)
        z = w
    return z

z = gradient_descent(zeta, 1/2+170*i,atol = 1e-40)
z
```

## Contour Integration

Cauchy's argument principle says that if $f$ is a meromorphic function on $\Omega$ whose boundary is the simple closed curve $\mathcal{C}$ (with $f$ holomorphic on $\mathcal{C}$), the number $Z$ of zeros of $f$ in $\Omega$ minus the number $P$ of poles of $f$ in $\Omega$ is counted by the integral
$$
    \frac{1}{2\pi i} \oint_{\mathcal{C}} \frac{f'(z)}{f(z)} = Z - P
$$
<img src="./images/Argument_principle.png"/>
<center> <font size =1> from wikimedia.org </font> </center>

```python
from scipy.integrate import quad

CC = ComplexField(prec = 100)
RR = RealField(prec = 100)

def complex_line_integral(f, init, final, N = 10000):
    z1 = CC(init)
    z2 = CC(final)
    real_f = lambda t: RR(f(z1 + RR(t)*(z2 - z1)).real_part())
    imag_f = lambda t: RR(f(z1 + RR(t)*(z2 - z1)).imag_part())
    
    real_int = quad(real_f, 0,1)[0]
    imag_int = quad(imag_f, 0,1)[0]
    
    return real_int*(z2 - z1) + CC(i)*imag_int*(z2-z1)

logderzeta = lambda s: zetaderiv(1,s)/zeta(s)


complex_line_integral(logderzeta, 3 ,50 + I)
```

```python
def complex_box_integral(f, left, right, bottom, top):
    bl = left + I*bottom
    br = right + I*bottom
    tl = left + I*top
    tr = right + I*top
    
    bint = complex_line_integral(f, bl, br)
    rint = complex_line_integral(f, br, tr)
    tint = complex_line_integral(f, tr, tl)
    lint = complex_line_integral(f, tl, bl)
    
    return (bint + rint + tint + lint)/(2*RR(pi)*CC(i))

complex_box_integral(logderzeta, -1,3,-1,50)
```

## Hardy Z function 

Note that we saw that the completed zeta function implied that on the $\operatorname{Re}(s) = 1/2$ line, functional equation
$$
\begin{align*}
        \zeta^*\left(\frac{1}{2} + it\right) &= \zeta^*\left(\frac{1}{2} - it\right) \\
        &= \zeta\left(\tfrac 12 - it\right) \Gamma\left(\tfrac{1}{4} + \tfrac{it}{2}\right) \pi^{-\tfrac{1}{4} + \tfrac{it}{2}} \\
        &= \overline{\zeta\left(\tfrac{1}{2} + it\right)} \overline{\Gamma\left(\tfrac{1}{4} + \tfrac{it}{2}\right) } \overline{\pi^{- \tfrac{1}{4} - \tfrac{it}{2}}}\\
        &= \overline{ \zeta^*\left(\frac{1}{2} + it\right) }
\end{align*}
$$
implies that $\zeta^*$ is real on the $1/2$-line. 

However since the $\Gamma$ function decays exponentially in the vertical direction, we wouldn't want to use $\zeta^*$ but only the phase of the extra terms.

Thus we define the Hardy Z-function to be
$$
    Z(t) = e^{i\vartheta(t)}\zeta\left(\frac{1}{2} + it \right)
$$
where $\vartheta(t)$ is the Riemann-Siegel theta function, 
$$
    \vartheta(t) = \arg \left(\Gamma\left(\frac{1}{4} + \frac{it}{2}\right)\right) - \frac{\log \pi}{2}t
$$

Notice that the zeros of the $Z(t)$ coincide with the zeros of a $\zeta(1/2 + it)$ for $t$ real. No new zeros are introduced.

```python
riemann_theta_asymp = lambda t : (t/2)*log(t/(2*RR(pi))) - t/2 - RR(pi)/8 + 1/(48*t) + 7/(5760*(t*t*t))
```

```python
RR = RealField(prec= 100)
CC = ComplexField(prec = 100)

HardyZ = lambda t: zeta(1/2 + i*RR(t))*(cos(riemann_theta_asymp(RR(t))) + CC(i)*sin(riemann_theta_asymp(RR(t))))
                                       
HardyZ(1000)
```

```python
HardyZreal = lambda t: HardyZ(t).real_part()

plot(HardyZreal,0,100)
```

```python
reset('RR')

def findZero(T,atol):
    """
    Returns the imaginary point of a zero of the Riemann Zeta Function
    zeta(1/2 + it) greater than T up to atol accuracy.
    """
    delta = 3/((log(RR(T + 10))))
    left = RR(T) + delta
    right = left + 2*delta
    while HardyZreal(left)*HardyZreal(right) > 0:
        left = right
        right = left + delta
    while delta > atol:
        delta = delta/2
        midpoint = left + delta
        if HardyZreal(left)*HardyZreal(midpoint) < 0:
            right = midpoint
        else:
            left = midpoint
    return left + delta/2

def findZeros(T, H, atol):
    """
    Returns a list of zeros of the Riemann zeta function 
    zeta(1/2 + it) on the interval t in [T, T+H] 
    accurate up to atol accuracy.
    """
    zero_list = []
    zero = findZero(T,atol)
    while zero < T + H:
        zero_list.append(zero)
        zero = findZero(zero, atol)
    return zero_list
    
findZeros(100,50,1e-6)
```

# Dirichlet L functions

There are other types of $L$ functions as well. 

For example take a Dirichlet character, these are multiplicative functions
$$
    \chi: (\mathbb{Z}/q\mathbb{Z})^\times \to S^1 \subset \mathbb{C}.
$$

This can be extended to $(\mathbb{Z}/q\mathbb{Z})$ by setting the terms to $0$, and then to $\mathbb{Z}$  via periodicity.

Because the group is finite the values are actually in the group of $\phi(N)$-th roots of unity. 

Their $L$-functions can also be defined similarly, and have Euler products, 
$$
    L(s,\chi) = \sum_{n=1}^\infty \frac{\chi(n)}{n^s} = \prod_{\text{prime } p} \left(1 - \frac{\chi(p)}{p^s}\right)^{-1}  .
$$

The completion of a Dirichlet $L$-function is 
$$
    L^*(s,\chi) = \begin{cases} q^{\frac{s}{2}}\pi^{-\frac{s}{2}} L(s,\chi)\Gamma\left(\frac{s}{2}\right) & \text{ if } \chi(-1) = 1\\
    q^{\frac{s}{2}} \pi^{- \frac{s}{2} } L(s,\chi) \Gamma\left(\frac{s+ 1}{2}\right) &\text{ if } \chi(-1) = -1.\end{cases}
$$

Since $q | q'$ implies that there are maps $\pi_{q', q} : \mathbb{Z}/q'\mathbb{Z} \longrightarrow \mathbb{Z}/q\mathbb{Z}$ we can pull a Dirichlet character modulo $q$ back to being a Dirichlet character modulo $q'$. 

If $\chi$ is a primitive Dirichlet character modulo $q$, that means if it doesn't come from a lower modulus, then the completed $L$-function also satisfies a functional equation
$$
    L^*(s,\chi)  = \varepsilon_\chi L^*(1-s, \overline{\chi})
$$
where $\varepsilon_\chi$ is a constant called the root number, satisfies $|\varepsilon_\chi| = 1$, and to be exact, is equal to
$$
    \varepsilon_\chi = \begin{cases} \tau(\chi)/\sqrt{q} & \text{ if } \chi(-1) = 1 \\
    i^{-1} \tau(\chi)/\sqrt{q} & \text{ if } \chi(-1) = -1\end{cases}
$$
where $\tau(\chi) = \sum_{a \mod q} \chi(n) e^{\frac{2\pi i a}{q}}$ is the Gauss sum.  For primitive $\chi$ modulo $q$, we have $|\tau(\chi)| = \sqrt{q}$, and $\overline{\varepsilon_\chi} = \varepsilon_{\overline{\chi}}$.

# Modular Forms

Dirichlet series are not the only ways to create $L$-functions. Another is modular forms.

A modular form of weight $k$ ($k$ even integer) on these are holomorphic functions on upper half plane $f:\mathbb{H} = \{x + iy \in \mathbb{C}: y >0 \}$ satisfying equations
$$
    f\left(\frac{az + b}{cz + d} \right) = (cz + d)^k f(z) \qquad \text{ for all} \begin{pmatrix} a& b  \\ c& d\end{pmatrix} \in \operatorname{SL}(2,\mathbb{Z})
$$
<img src ="./images/fundamentalDomain.png"/>

The transformation via $\left(\begin{smallmatrix} 1 & 1\\ 0 & 1 \end{smallmatrix} \right)$ forces $f(z + 1) = f(z)$ and therefore 
$$
f(z) = \sum_{n =1}^\infty  \lambda_f(n) n^{\frac{k-1}{2}} e^{2\pi i n z}.
$$

The $L$-function of a Hecke modular form $f$ is given by
$$
       L(s,f) = \sum_{n=1}^\infty \frac{\lambda_f(n)}{n^s} = \prod_{p \text{ prime}} \left(1-\frac{\lambda_f(p)}{p^s} + \frac{1}{p^{2s}}\right)^{-1} = \prod_{p \text{ prime }} \left(1 - \frac{\alpha_f(p)}{p^s}\right)^{-1} \left(1 - \frac{\beta_f(p)}{p^s}\right)^{-1}.
$$

Its completion is given by
$$
    L^*(s,f) = \pi^{-s} L(s,f) \Gamma\left(\frac{s + \frac{k-1}{2}}{2}\right) \Gamma\left(\frac{s + \frac{k+1}{2}}{2}\right)
$$
and satisfies the functional equation
$$
    L^*(s,f) = i^{-k} L^*(1-s, \overline{f}).
$$

# Half-integer weight modular forms.

This part is joint work with David <i>Thomas Hulse, Li Mei Lim and David Lowry-Duda</i>. We started this project in Nesin Matematik Köyü in the summer of 2018.

<img src="./images/Turkey-ConferencePhoto.jpg" width = 700px/>


Similar to full integer weight modular forms there is another class of holomorphic functions, which transform with a factor of automorphy that is $(c + dz)^k$ for a half integer $k$.

More precisely they are $f$ such that for an odd $k$,
$$
    f\left(\frac{az + b}{cz + d} \right) = \varepsilon_d^{-1} \left(\frac{c}{d}\right) (cz + d)^{k/2} f(z) \qquad \qquad \text{ for all } \begin{pmatrix}a&b\\ 4c & d\end{pmatrix} \in \operatorname{SL}(2, \mathbb{Z}), a,b,c,d \in \mathbb{Z}
$$
where $\left(\frac{c}{d}\right)$ is the binary Kronecker symbol, $\varepsilon_d = \begin{cases}1 &\text{ if } d \equiv 1 \pmod 4\\i &\text{ if } d \equiv 3 \pmod 4\end{cases}$, and $k$ is an odd integer.

There are also Hecke operators on this space.


These $L$-functions also have completions, analytic continuations, a functional equation all very similar to the previous cases...


But no EULER PRODUCT!

Lo and behold, there are zeros off the critical line.
